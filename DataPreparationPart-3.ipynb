```python
import pandas as pd
```

```python
fp1=pd.read_csv('CrimeData_Cleaned.csv')
fp2=pd.read_csv('BusinessData_Cleaned.csv')
fp3=pd.read_csv('SchoolsData_Cleaned.csv')
```

```python
fp3['ZIP CODE']=fp3['ZIP Code']
fp3=fp3.drop(columns=['ZIP Code'])
```

```python
fp_integrated = pd.merge(fp2,fp3,on=['ZIP CODE'],how='inner')
```

```python
fp_integrated.info()
```

```python
fp_integrated['Year']=fp_integrated['Year_x']
fp_integrated['Neighbourhood']=fp_integrated['Neighbourhood_x']
fp_integrated=fp_integrated.drop(columns=['Year_x','Neighbourhood_x'])
fp_integrated=fp_integrated.drop(columns=['Year_y','Neighbourhood_y'])
```

```python
fp_integrated1 = pd.merge(fp1,fp_integrated,on=['Neighbourhood'],how='inner')
```

```python
fp_integrated1
```

```python
fp_integrated1.info()
```

```python
fp_integrated1.to_csv('Integrated_Dataset.csv',index=None)
```
